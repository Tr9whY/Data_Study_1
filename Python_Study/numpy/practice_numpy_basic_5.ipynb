```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9,4))

np.random.seed(100)

all_data = [np.random.normal(0, std, 100) for std in range(6,10)]

# violin plot
axs[0].violinplot(all_data, showmeans = False, showmedians =True)
axs[0].set_title('Violin plot')

# box plot
axs[1].boxplot(all_data)
axs[1].set_title('Box plot')

for ax in axs:
    ax.yaxis.grid(True)
    ax.set_xticks([i+1 for i in range(len(all_data))], labels=['x1','x2','x3','x4'])
    ax.set_xlabel('Four separated samples')
    ax.set_ylabel('observed values')
    
    # grid : 격자
    # tick : 눈금자
    # label : 라벨, 명칭
    # legend : 범례
plt.show()
```

### HISTOGRAM 히스토그램

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

numbers = np.random.normal(size=10000)
print(numbers)          # 연속형 데이터

plt.hist(numbers, color='#7f00ff')

plt.xlabel('value')
plt.ylabel('freq')

plt.show()                  # 가운데를 기준으로 선을 그리면 정규분포
```

```python
data = np.random.randn(1000)

#히스토그램 그리기
plt.hist(data, bins = 10, color = '#7f00ff') # 계급 구간
plt.title('histogram')      # 제목

plt.xlabel('value')
plt.ylabel('freq')

plt.grid(True, color = 'gray')
plt.show()
```

```python
# 지수 함수
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,10)

y1 = np.ones(10)
y2 = x
y3 = x**2

# 잘 분리해줄 것..
plt.plot(x, y1,label='y=1' ,color='Darkblue')
plt.plot(x, y2,label='y=x' ,color='Darkred')
plt.plot(x, y3,label='y=x^2' ,color='#7f00ff')

plt.legend()
plt.show()

# plt.plot(x, y1,y2,y3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def sigmoid(x):
    s = 1 / (1+np.exp(-x))      # sigmoid (시그모이드) 공식
    ds = s*(1-s)
    return s, ds

x = np.linspace(-10, 10 , 100)  # 엔빵
y1 , y2 = sigmoid(x)

# 잘 분리해줄 것..
#plt.plot(x, y1)
#plt.plot(x, y2)
plt.plot(x, y1, x, y2)

plt.xlabel('x')
plt.ylabel('sigmoid(x), sigmoid_prime(x)')

plt.show()
```
