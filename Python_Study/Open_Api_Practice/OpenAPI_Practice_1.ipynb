```python
import requests
import xml.etree.ElementTree as ET

# 발급받은 API Key
api_key = ''

# 사용할 API URL 설정
api_url = ''

# 요청 파라미터 설정
params = {
  'serviceKey' : api_key,
  'service_Type' : 'json', #제공하는 파일형식(xml)
  'Page_No' : '1',
  'Page_Size' : '20',
  'food_Name' : '밥' #검색하고자하는 음식명(옵션)
}

# API 요청 보내기
response = requests.get(api_url, params=params)

#데이터 값 출력해보기
contents = response.text
```

```python
data = response.json()
print(json.dumps(data, indent=4, ensure_ascii=False))
```

```python
print(data['response'])
print(data['response']['list'])
print(data['response']['list'][0])
print(data['response']['list'][0]['food_List'])
print(data['response']['list'][0]['food_List'][0])
print(data['response']['list'][0]['food_List'][0]['food_Nm'])
```

```python
if response.status_code == 200:
    try:
        data=response.json()
        print(json.dumps(data, indent=4, ensure_ascii=False))
        
        items = data['response']['list']
        parsed_data = []
        
        for item in items:
            parsed_data.append({
                'fd_Code': item['fd_Code'],
                'upper_Fd_Grupp_Nm' : item['upper_Fd_Grupp_Nm'],
                'fd_Grupp_Nm' : item['fd_Grupp_Nm'],
                'fd_Nm' : item['fd_Nm'],
                'fd_Wgh' : item['fd_Wgh'],
                'food_Cnt':item['food_Cnt'],
                'food_List' : item['food_List']
                })
            # 추출한 데이터를 판다스 데이터프레임으로 변환
            df = pd.DataFrame(parsed_data)
            print(df)
    except json.JSONDecodeError:
        print("Error 발생")
```
