```python
with open(input+'.json', 'r', encoding='utf-8') as file:
  data = json.load(file)

print(data)
```

```python
message = ""



for item in data:
  # print(item)
      if 'message' in item.keys():
          new_message = re.sub(r'[^\w]','', item['message'])

          message += new_message
          print(new_message) # 전처리 된 새로운 메세지
          #print(message) # 각 단계별 message 변화 출력 (업데이트 됨)

print(message)
```

```python
message = ""

seen_messages = set()  # 중복 메세지 추적하기 위한 집합

for item in data:
  # print(item)
      if 'message' in item.keys():
          new_message = re.sub(r'[^\w]','', item['message'])
          if new_message not in seen_messages:
            seen_messages.add(new_message)
            message += new_message
          #print(new_message)

print(message)
```

```python
message
```

```python
# 형태소 분류기
nlp = Okt()
print(nlp.nouns(message)) #noun : 명사
```

```python
message_n = nlp.nouns(message)
```

```python
count_n = Counter(message_n)
count_n
```

```python
word_count = dict()

for tag, counts in count_n.most_common(80):
    if (len(str(tag)) > 1):
        word_count[tag] = counts
        print("%s, %d"%(tag, counts))
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
font = 'NanumBarunGothic'
```

```python
plt.figure(figsize=(12,6))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid()

sorted_keys = sorted(word_count, key=word_count.get, reverse=True )
sorted_values= sorted(word_count.values(), reverse=True)

plt.bar(range(len(word_count)), sorted_values, align='center')
plt.xticks(range(len(word_count)), list(sorted_keys), rotation=80)
plt.show()
```

```python
from wordcloud import WordCloud

wc = WordCloud(font, background_color='ivory', width = 800, height= 600)
wc.generate_from_frequencies(word_count)

plt.figure(figsize=(8,8))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
