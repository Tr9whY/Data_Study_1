```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA

from sklearn import datasets, linear_model

from sklearn.metrics import mean_squared_error, r2_score
```

```python
path = '/content/Advertising.csv'

pd.read_csv(path, index_col=0)

ad = pd.read_csv(path, index_col=0)
ad
```

```python
# pcr 모델 적용

# 독립변수(x), 종속변수(y) 분리

ad.columns
```

```python
x = ad[['TV', 'Radio', 'Newspaper']]
y = ad[['Sales']]
```

```python
x
```

```python
x.shape
```

```python
y
```

```python
y.shape
```

```python
# 주요인분석 수행
# (features 3 >> n_components 2 설정)
pca = PCA(n_components=2)
```

```python
# 적합(fitting)
pca.fit(x)
```

```python
pca.explained_variance_ratio_
# 2개 변수 >> 97.7% 설명됨
```

```python
# PC score
z = pca.fit_transform(x)
z
```

```python
# PCA 한 후 scaled 된 값 활용
# >> 훈련용 데이터, 테스트 데이터 구분
from sklearn.model_selection import train_test_split

z_train, z_test, y_train, y_test =\
train_test_split(z, y, test_size=0.2, random_state=42)
```

```python
z_train[:5]
```

```python
z_train.shape, z_test.shape, y_train.shape, y_test.shape
```

```python
# 선형회귀 객체 생성
lr = linear_model.LinearRegression()

lr.fit(z_train, y_train)
```

```python
# 예측

# 1. training data 활용, 예측
z_train_pred = lr.predict(z_train)
# 2. test data 활용, 예측 (**)
z_test_pred = lr.predict(z_test)
```

```python
z_train_pred[:5]
```

```python
y_train[:5]
```

```python
z_test_pred[:5]
```

```python
y_test[:5]
```

```python
# coefficients
print("coefficients: ", lr.coef_)
```

```python
# train data MSE
print("Training data MSE: %.3f"%mean_squared_error(y_train, z_train_pred))
```

```python
# test data MSE (**)
print("Test data MSE: %.3f"%mean_squared_error(y_test, z_test_pred))
```

```python
# 결정계수
print("Training r2_score: %.3f"%r2_score(y_train, z_train_pred))
print("Test r2_score: %.3f"%r2_score(y_test, z_test_pred)) #(**)
```
