### 지도학습 (분류)

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
```

```python
data.data
```

```python
data.target
```

```python
x = data.data
y = data.target
```

```python
x = x[:, :10]
```

```python
# 로지스틱 회귀 (분류)
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='lbfgs')
```

```python
model.fit(x, y)
```

```python
# 모델 >> 예측

model.predict(x)

y_pred = model.predict(x)
y_pred[:10]
```

```python
# 정확도 측정

from sklearn.metrics import accuracy_score # 정확도 점수를 매겨주는 것

accuracy_score(y, y_pred)
```

### 비지도 학습 예

```python
from sklearn.datasets import load_wine

data = load_wine()
```

```python
x = data.data[:,[0,9]]
```

```python
from sklearn.cluster import KMeans

n_cluster = 3
model = KMeans(n_clusters=n_cluster)
```

```python
pred = model.fit_predict(x)
```

```python
pred
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(x[pred==0, 0], x[pred==0, 1], color='red', marker='o', label='Label1')
ax.scatter(x[pred==1, 0], x[pred==1, 1], color='blue', marker='o', label='Label2')
ax.scatter(x[pred==2, 0], x[pred==2, 1], color='green', marker='o', label='Label3')
ax.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=200, color='yellow', marker="*", label="center")

ax.legend()
plt.show()
```
