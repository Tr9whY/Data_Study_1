Matplotlib 주요 그래프

```python
import matplotlib.pyplot as plt
import numpy as np

# fig, ax = plt.subplot() -> 올바른 코드는 아래와 같습니다.
fig, ax = plt.subplots()

x = np.arange(15)
y = x ** 2

ax.plot(
    x, y,
    linestyle=':',  # 선 스타일을 지정합니다.
    marker='*',     # 마커를 지정합니다.
    color='darkblue'  # 선 색상을 지정합니다.
)
plt.show()
```

```python
# color

x = np.arange(10)

fig, ax = plt.subplots()

ax.plot(x, x ,color ='r')
ax.plot(x, x+2, color = 'green')

#그레이스케일 (흑백) 0~1 
ax.plot(x, x+4, color = '0.8')
ax.plot(x , x+6 , color = '#7f00ff')

plt.show()
```

```python
# marker

x = np.arange(10)
fig,ax = plt.subplots()
ax.plot(x, x, marker='.')
ax.plot(x, x+2, marker='v')
ax.plot(x, x+4, marker = 's')
ax.plot(x, x+4, marker = '*')

plt.show()
```

```python
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots()

ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12)
ax.set_ylim(-2, 4)

plt.show()
```

```python
# 범례 (legend)

fig,ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')

ax.set_title("legend of graph")
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.legend(loc = 'best', shadow=False, fancybox=False, borderpad=2)

plt.show()
```

```python
# 산점도 (scatter plot...중요!!!!!!)

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)

fig, ax = plt.subplots()
ax.scatter(
    x, x**2,            # x와 y 데이터
    marker='*',         # 마커 모양
    s=50,               # 마커 크기
    facecolors='white', # 마커 내부 색상
    edgecolors='blue'   # 마커 테두리 색상
)

plt.show()
```

```python
# bubble 형태 scatter    ### 중요!

fig, ax =plt.subplots()

x = np.random.randn(50)
y = np.random.randn(50)

colors = np.random.randint(0, 100, 50)
sizes= 500 * np.pi * np.random.rand(50) ** 2

ax.scatter(x, y, c = colors, s = sizes, alpha = 0.25)  #alpha = 투명도

plt.show()
```

```python
#바 그래프 (bar plot)

# bar

x= np.arange(10)
fig,ax = plt.subplots(figsize=(12,4))
ax.bar(x, x*2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)

data = [x, y, z]
fig, ax = plt.subplots()

x_ax = np.arange(3)

for i in x_ax:
    ax.bar(x_ax, data[i], bottom=np.sum(data[:i], axis=0))

ax.set_xticks(x_ax)
ax.set_xticklabels(['A', 'B', 'C'])

plt.show()
```

```python
# histogram(히스토그램)

fig, ax = plt.subplots()
data = np.random.randn(1000)

ax.hist(data, bins = 50) # bins : 계급구간

plt.show()
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(df['order'], df['height(cm)'], label='height')

ax.set_xlabel('order')
ax.set_ylabel('height(cm)')
```
