```python
import matplotlib.pyplot as plt
import numpy as np

# fig, ax = plt.subplot() ->    .
fig, ax = plt.subplots()

x = np.arange(15)
y = x ** 2

ax.plot(
    x, y,
    linestyle=':',  #   .
    marker='*',     #  .
    color='darkblue'  #   .
)
plt.show()
```

```python
# color

x = np.arange(10)

fig, ax = plt.subplots()

ax.plot(x, x ,color ='r')
ax.plot(x, x+2, color = 'green')

# () 0~1 
ax.plot(x, x+4, color = '0.8')
ax.plot(x , x+6 , color = '#7f00ff')

plt.show()
```

```python
# marker

x = np.arange(10)
fig,ax = plt.subplots()
ax.plot(x, x, marker='.')
ax.plot(x, x+2, marker='v')
ax.plot(x, x+4, marker = 's')
ax.plot(x, x+4, marker = '*')

plt.show()
```

```python
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots()

ax.plot(x, np.sin(x))
ax.set_xlim(-2, 12)
ax.set_ylim(-2, 4)

plt.show()
```

```python
#  (legend)

fig,ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')

ax.set_title("legend of graph")
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.legend(loc = 'best', shadow=False, fancybox=False, borderpad=2)

plt.show()
```

```python
#  (scatter plot...!!!!!!)

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)

fig, ax = plt.subplots()
ax.scatter(
    x, x**2,            # x y 
    marker='*',         #  
    s=50,               #  
    facecolors='white', #   
    edgecolors='blue'   #   
)

plt.show()
```

```python
# bubble  scatter    ### !

fig, ax =plt.subplots()

x = np.random.randn(50)
y = np.random.randn(50)

colors = np.random.randint(0, 100, 50)
sizes= 500 * np.pi * np.random.rand(50) ** 2

ax.scatter(x, y, c = colors, s = sizes, alpha = 0.25)  #alpha = 

plt.show()
```

```python
#  (bar plot)

# bar

x= np.arange(10)
fig,ax = plt.subplots(figsize=(12,4))
ax.bar(x, x*2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)

data = [x, y, z]
fig, ax = plt.subplots()

x_ax = np.arange(3)

for i in x_ax:
    ax.bar(x_ax, data[i], bottom=np.sum(data[:i], axis=0))

ax.set_xticks(x_ax)
ax.set_xticklabels(['A', 'B', 'C'])

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'president_heights.csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(df['order'], df['height(cm)'], label='height')

ax.set_xlabel('order')
ax.set_ylabel('height(cm)')
```

```python
condition_01 = (df1['Type 1'] == 'Fire') | (df1['Type 2'] == 'Fire')

df1[condition_01]
```

```python
water
```

```python
fig, ax = plt.subplots()

ax.scatter(fire['Attack'], fire['Defense'], color='red', label='Fire', marker='o', s=50)

ax.scatter(water['Attack'], water['Defense'], color='blue', label='Fire', marker='o', s=30)

ax.set_xlabel('Attack')
ax.set_ylabel('Defense')
ax.legend(loc='best')

plt.show()
```
