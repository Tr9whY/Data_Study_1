```python
import pandas as pd

path = '송파구_동별반려동물등록현황.csv'
df = pd.read_csv(path, encoding = 'utf-8')

# deep copy
sample = df.copy()

# column 정리
col = ['자치구명','법정동명','등록동물수(마리)','등록소유자수']

sample_chose_col = sample[col]
sample_chose_col.head(2)
```

```python
sample_rename_col = sample_chose_col.rename(columns={'자치구명' : 'Gu',
                                                     '법정동명':'Dong',
                                                     '등록소유자수':'Pet_Owners',
                                                     '등록동물수(마리)':'Total_Registrations'})
sample_rename_col.head()
```

```python
sample_rename_col.shape
```

```python
check_dup = sample_rename_col.copy()
```

```python
check_dup = check_dup.drop_duplicates(subset=['Gu','Dong'], keep='first').reset_index(drop=True)
check_dup.shape
```

```python
Songpa_Pet = check_dup.copy()
```

```python
Songpa_Pet = Songpa_Pet.sort_values(by=['Gu','Dong'])
```

```python
Songpa_Pet['Idx'] = range(len(Songpa_Pet))
Songpa_Pet.info()
```

```python
Col_sort = ['Idx','Gu','Dong','Pet_Owners','Total_Registrations']
Songpa_Pet = Songpa_Pet[Col_sort]
Songpa_Pet.head(3)
```

```python
Songpa_Pet.to_csv('Gu_Dong_Songpa_Pet.csv', encoding='utf-8')
```

```python
Songpa_Pet_Owners_r = Songpa_Pet['Total_Registrations'].describe() / Songpa_Pet['Pet_Owners'].describe()
```
